### Keyword analysis
- Identify core keywords among the keywords selected by the authors based on several indicators (e.g. frequency, various centrality ect.).
- Because the journal "Environmental History" does not have author keywords, index keywords are used.
- All data were exported from Scopus and are in CSV format.

```python
import sys
from os import getcwd
from os.path import abspath, dirname, join
sys.path.append(f'{getcwd()}/bibliometric')
import warnings
warnings.simplefilter('ignore', category=(UserWarning, FutureWarning, SyntaxWarning))

import pandas as pd
pd.set_option('mode.chained_assignment', None)

from bibliometric import tools

BASE_PATH = abspath(getcwd())
PARENT_PATH = dirname(dirname(BASE_PATH))
RESOURCE_PATH = join(PARENT_PATH, 'resources')
target_file_path = join(RESOURCE_PATH, '20250711_scopus_work2.csv')
```

#### Load the CSV file of the journals.

```python
df = pd.read_csv(target_file_path, encoding='utf-8', sep=',')
```

#### Check highly counted keywords.

Top 10 highly counted keywords for the entire period.

```python
tools.plot_top_(df, category='author keyword', top_num=10)
```

```python
tools.plot_top_(df, category='index keyword', top_num=10)
```

```python
tools.plot_top_(df, category='keyword', top_num=10)
```
