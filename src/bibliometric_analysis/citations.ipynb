### Citation analysis
- Identify significant cited authors and articles based on several indicators (e.g. frequency, various centrality ect.).
- All data were exported from Scopus and are in CSV format.

```python
import sys
from os import getcwd
from os.path import abspath, dirname, join
sys.path.append(f'{getcwd()}/bibliometric')
import warnings
warnings.simplefilter('ignore', category=(UserWarning, FutureWarning, SyntaxWarning))

import pandas as pd
pd.set_option('mode.chained_assignment', None)

from bibliometric import tools

BASE_PATH = abspath(getcwd())
PARENT_PATH = dirname(dirname(BASE_PATH))
RESOURCE_PATH = join(PARENT_PATH, 'resources')
target_file_path = join(RESOURCE_PATH, '20250711_scopus_work2.csv')
```

#### Load the CSV file of the jounals.

```python
df = pd.read_csv(target_file_path, encoding='utf-8', sep=',')
```

#### Check average number of references per pubication by year.

```python
tools.plot_references(df=df)
```

#### Check highly counted references

Top 10 highly counted references for the entire period.

```python
tools.plot_top_(df=df, category='reference', top_num=10)
```

#### Check references with high centrality among reference relationship.

Top 10 references with high PageRank centrality for the entire period.

```python
tools.get_directed_graph_centrality_(df=df, centrality='pagerank', top_num=10)
```

#### Check highly referenced authors.

Top 10 highly referenced authors for the entire period.

```python
tools.plot_top_authors_of_references(df=df, delimiter=',', top_num=10)
```

#### Check descriptive statistics of references.

Descriptive statistics and age of references with upper limit of 50 for the entire period.

```python
tools.plot_age_of_citataions(df=df, top_age=50)
```
